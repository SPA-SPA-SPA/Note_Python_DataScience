### 代码段计时：`%timeit`和`%time`

`%%timeit`可以让代码段重复运行来计算代码的运行时间

````python
%timeit sum(range(100))
````

````python
range?
````

````python
%%timeit
total = 0
for i in range(1000):
    for j in range(1000):
        total += i * (-1) ** j
````

对于较慢的命令，`%timeit`会自动调整并减少重复执行的次数。

有时候重复一个操作不是最佳的选择。例如，如果有一个列表需要排序，我们可能会被重复操作误导。对于一个预先排好序的列表进行排序，比对一个无序的列表进行排序要快，所以重复运行将使结果出现偏差：

````python
import random
L = [random.random() for i in range(100000)]
%timeit L.sort()
````

对于这种情况，使用`%time`魔法函数可能是更好的选择，下面对一个无序列表排序和一个已排序的列表排序分别计时：

````python
import random
L = [random.random() for i in range(100000)]
print("sorting an unsorted list:")
%time L.sort()
````

````python
print("sorting an already sorted list:")
%time L.sort()
````

可以看到，`%timeit`计时比`%time`要快，那是因为前者会阻止“垃圾回收”，这缩短了时间。

### 分析整个脚本：`%prun`

有时候要对整个脚本计时，python包含一个内置的代码分析器，但是IPython提供了一种更加方便的方式来使用这个分析器，也就是通过魔法函数`%prun`实现。

下面例子，我们定义一个简单的函数，然后用`%prun`和一个函数调用来看分析结果。

````python
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j >> i) for j in range(N)]
        total += sum(L)
    return total
````

````python
%prun sum_of_lists(1000000)
````

结果是一个表格，按照每个函数调用的总时间，显示哪里的执行时间最长。在这个例子中，大部分执行时间用在sum_of_lists的列表综合中。通过观察这个数据，我们可以考虑通过调整哪里来提升算法的性能。

使用?获得更多的信息。

### 用`%memit`和`%mprun`进行内存分析

另一种分析是分析一个操作的所用内存量，这个可以通过IPython的另一个扩展来评估，也即memory_profiler。安装过程和1.9.3一样。


````python
%load_ext memory_profiler
````

这个内存分析扩展包中有两个魔法函数：
1. `%memit`：提供的内存消耗计算功能类似于`%timeit`
1. `%mprun`：提供的内存消耗计算功能类似于`%lprun`

`%memit`的使用很简单：

````python
%memit sum_of_lists(1000000)
````

对于逐行代码的内存消耗描述，可以使用`%mprun`魔法函数。但是不幸的是，这个魔法函数仅仅对独立模块内部的函数有效，而对Notebook本身不起作用。所以首先使用`%%file`魔法函数创建一个简单的模块，将该模块命名为`mprun_demo.py`，它包含sum_of_lists函数，该函数中包含一次加法，能使内存分析结果更加清晰：

````python
%%file mprun_demo.py
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j >> i) for j in range(N)]
        total += sum(L)
        del L # remove reference to L
    return total
````

````python
from mprun_demo import sum_of_lists
%mprun -f sum_of_lists sum_of_lists(1000000)
````

页面中打印的结果概述了该函数的内存消耗情况。其中，Increment列告诉我们每行代码对总内存预算的影响：创建和删除列表L时用掉了25MB内存。这是除了python解释器本身外最消耗内存资源的部分。

## 理解Python中的数据类型

首先你要理解数据是如何被存储和操作的。本节将介绍Python语言中数据数组是如何被处理的，并对比NumPy所作的改进。**理解这个而不同之处正是理解本书其他内容的基础。**

Python是动态类型的语言，这是在说：你可以不用规定一个变量的数据类型。为什么python可以这么做，而C语言不可以？理解这个特性如何工作十分重要！

### python整型不仅仅是一个整型

**标准的python实现是用C语言编写的。** 这意味着每一个python对象都是一个聪明的伪C语言结构体，该结构体不仅包含其值，还有其他信息。例如，当我们在python中定义一个整型，例如`x = 10000`时，x 并不是一个“原生”整型，而是一个指针，指向一个C语言的复合结构体，结构体里包含了一些值。查看Python的源代码，可以发现整型（长整型）的定义，如下定义（C语言的宏经过扩展之后）：


```C
struct _longobject{
    # 引用计数，它帮助python默默地处理内存的分配和回收
    long ob_refcnt;
    
    # 将变量的类型编码
    PyTypeObject *ob_type;
    
    # 指定接下来的数据成员的大小
    size_t ob_size;
    
    # 包含我们的希望python变量表示的实际整型值
    long ob_digit[1];
};
```

两者的差异在于，C语言整型本质上时对应某个内存位置的标签，里面存储的字节会编码成整型。而python的整型其实是一个指针，指向包含这个python对象所有信息的某个内存位置，其中包括可以转换成整型的字节。由于python的整型结构体里面还包含了大量额外的信息，所以python可以自由、动态地编码。但是，python类型中的这些额外信息也会成为负担，在多个对象组合的结构体中尤其明显。

### python列表不仅仅是一个列表

python中的标准可变多元素容器是列表。你可以创建一个整型值列表，也可以创建一个字符串列表，还可以创建一个异构的列表：

````python
L = list(range(10))
L2 = [str(c) for c in L]
L3 = [True, "2", 3.0, 4]

print(L)
print(type(L[0]))

print(L2)
print(type(L2[0]))

print(L3)
print([type(item) for item in L3])
````

> 记住：列表是方括号，或者说是中括号：\[\]。

现在看一个例子，如果列表中的所有类型都是同一个类型的，那么很多信息都会显得多余——将数据存储在固定类型的数组中会更高效。这里要注意图2-2：C列表和python列表的区别。

## NumPy数组的计算：通用函数

前面介绍了一些NumPy的一些基础知识，接下来深入了解为什么NumPy在Python数据科学世界中如此重要的原因，明确点说，NumPy提供了一个简单灵活的接口来优化数据数组的计算。

NumPy数组的计算时快时慢。使NumPy变快的关键是利用 **向量化** 操作，通常在NumPy的 **通用函数(ufunc)** 中实现。本节会介绍NumPy通用函数的重要性——它们可以提高数组元素的重复计算的效率。

### 缓慢的循环

python的默认实现（被称作CPython）有时候很慢，这是因为该语言的动态性和解释性。所以由很多项目想把python代码转化为可编译的C语言代码。

一般来说，循环的时候是较缓慢的。比如有一个数组，我们想计算每个元素的倒数，一种直接的解决方法是：

````python
import numpy as np
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

values = np.random.randint(1, 10, size=5)
compute_reciprocals(values)
````

这种实现方式可能对于有C语言背景的人来说非常自然，但是如果测试一个很大的输入数据运行上述代码的时间，会发现这个操作很耗时。我们用IPython的`%timeit`魔法函数来测量：

````python
big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
````

完成百万次上述操作并存储结果居然花了几秒钟的时间！！！现在连手机都能以每秒十亿次浮点计算的速度进行处理了，上面的操作的确是不合时宜的慢。

但事实上，这里的处理瓶颈并不是运算本身，而是CPython在每次循环时必须做数据类型的检查和函数的调度。每次进行倒数运算时，python首先检查对象的类型，并且动态查找可以使用该数据类型的正确函数。

如果我们在编译代码的时候就进行这样的操作，那么就能在代码执行之前知晓类型的声明了，结果的计算也会更加有效率。

### 通用函数介绍

NumPy为很多类型的操作提供了非常方便的、静态类型的、可编译程序的接口，也被称作 **向量操作**。这里对数组的操作将会被作用于数组中的每一个元素。这种向量方法被用于将循环推送至NumPy之下的编译层，这样会取得更快的执行效率。

比较以下两个结果：

````python
print(compute_reciprocals(values))
print(1.0 / values)
````

如果计算一个较大数组的运行时间，可以看到它的完成时间比Python循环花费的时间更短：

````python
%timeit (1.0 / big_array)
````

这是因为NumPy中的向量操作时通过通用函数实现的，通用函数的主要目的时对NumPy数组中的值执行更快的重复操作。不仅可以进行标量和数组的运算，也可以对两个数组进行运算。

````python
np.arange(5) / np.arange(1, 6)
````

通用函数并不仅限于一维数组的运算，它们也可以进行多维数组的运算：

````python
x = np.arange(9).reshape((3, 3))
2 ** x
````

通过通用函数，用向量的方式计算几乎总是比用python循环实现的计算更加有效，尤其是当数组很大时。只要你看到python脚本中有这样的循环，就应该考虑能否用向量方式替换这个循环。

很多NumPy用户在没有完全了解通用函数的特征时就开始使用它们，这里介绍一些通用函数的特殊性质。

#### 指定输出

在进行大量运算时，有时候指定一个用于存放运算结果的数组时非常有用的。不同于创建临时数组，你可以用这个特性将计算结果直接写入到你期望的存储位置。所有的通用函数都可以通过out参数来指定计算结果的存放位置：

````python
x = np.arange(5)
y = np.empty(5)
np.multiply(x, 10, out=y)
print(y)
````

这个特性也可以被用作数组视图，例如可以将计算结果写入指定数组的每隔一个元素的位置：

````python
y = np.zeros(10)
np.power(2, x, out=y[::2])
print(y)
````

如果这里写的是`y[::2] = 2 ** x`，那么结果将是创建一个临时数组，该数组存放的是`2 ** x`的结果，并且会将这些指复制到y数组中。对于上述例子中比较小的计算量来说，这两种方式的差别不大。但是对于较大的数组，通过慎重使用out参数将能够有效节约内存。

#### 聚合

二元通用函数有些非常有趣的聚合功能，这些聚合可以直接在对象上计算。例如，如果我们希望用一个特定运算`reduce`一个数组，那么可以用任何通用函数的`reduce`方法。一个`reduce`方法会对给定的元素和操作重复执行，直至得到单个的结果。

例如，对`add`通用函数调用`reduce`方法会返回数组中的所有元素的和：

````python
x = np.arange(1, 6)
np.add.reduce(x)
````

同样，对multiply通用函数调用reduce方法会返回数组中所有元素的乘积：

````python
np.multiply.reduce(x)
````

如果需要存储每次计算的中间结果，可以使用accumulate:

````python
np.add.accumulate(x)
````

请注意，在一些特殊的情况中，NumPy提供了专用的函数(np.sum, np.prod, np.cumsum, np.cumprod)，它们也可以实现以上reduce的功能，这些函数将在2.4节中具体介绍。

#### 外积

最后，任何通用函数都可以用outer方法获得两个不同输入数组所有元素对的函数运算结果。这意味着你可以用一行代码实现一个乘法表：

````python
x = np.arange(1, 6)
np.multiply.outer(x, x)
````

````python
np.add.outer(x, x)
````

## 数组的计算：广播

还有一种向量化操作的方法是利用NumPy的广播功能。广播可以简单理解为用于不同大小数组的二进制通用函数（加、减、乘等）的一组规则。

### 广播的介绍

前面曾提到，对于同样大小的数组，二元操作是对相应元素逐个计算：

````python
import numpy as np
a = np.array([0, 1, 2])
b = np.array([5, 5, 5])
a + b
````

广播允许这些二元操作可以用于不同大小的数组。例如，可以简单地将一个标量（可以认为是一个零维的数组）和一个数组相加：

````python
a + 5
````

NumPy广播功能的好处是，这种对值得重复实际上并没有发生，但是这是一种很好用得理解广播的模型。

我们同样也可以将这个原理扩展到更高维度的数组，观察以下将一个一维数组和一个二维数组相加的结果：

````python
M = np.ones((3,3))
M
````

````python
M + a
````

这里这个一维数组就被扩展或者广播了，它沿着第二个维度扩展，扩展到匹配M数组的形状。

以上的这些例子理解起来都比较容易，更复杂的情况会涉及对两个数组的同时广播，例如：

````python
a = np.arange(3)
b = np.arange(3)[:, np.newaxis]
print(a)
print(b)
````

````python
a + b
````

以上例子的几何可视化如下图：

NumPy广播的可视化

浅色的盒子表示广播的值。同样需要注意的是，这个额外的内存并没有在实际操作上进行分配，但是这样的想象方式更方便我们从概念上理解。

### 广播的规则

NumPy的广播遵循一组严格的规则，设定这组规则是为了决定两个数组间的操作。

1. 如果2个数组的维度数不同，那么最小维度数组的形状会在最左边补1。
1. 如果2个数组的形状在任何一个维度都不匹配，那么数组的形状会沿着维度为1的维度扩展以匹配另一个数组的形状。
1. 如果2个数组的形状在任何一个维度上都不匹配并且没有任何一个维度等于1，那么会引起异常。

#### 广播示例1

将一个二维数组与一个一维数组相加：

````python
M = np.ones((2, 3))
a = np.arange(3)
print(M.shape)
print(a.shape)
````

根据规则1，数组a的维度更小，所以在其左边补1：

```python
M.shape -> (2, 3)
a.shape -> (1, 3)
```

根据规则2，第一个维度不匹配，因此扩展这个维度以匹配数组：

```python
M.shape -> (2, 3)
a.shape -> (2, 3)
```

现在两个数组的形状匹配了，可以看到它们的最终形状都为(2, 3)：

````python
M + a
````

#### 广播示例2

来看2个数组均需要广播的示例：

````python
a = np.arange(3).reshape((3, 1))
b = np.arange(3)
````

同样，首先写出2个数组的形状：

```python
a.shape = (3, 1)
b.shape = (3,)
```

规则1告诉我们，需要用1将b形状补全：

```python
a.shape -> (3, 1)
b.shape -> (1, 3)
```

规则2告诉我们，需要更新这两个数组的维度来互相匹配：

```python
a.shape -> (3, 3)
b.shape -> (3, 3)
```

````python
a + b
````

#### 广播示例3

现在来看一个两个数组不兼容的示例：

````python
M = np.ones((3, 2))
a = np.arange(3)
````

和第一个示例相比，这里有个微小的不同之处：矩阵M是转置的。那么这将如何影响计算呢？两个数组的形状如下：

```python
M.shape = (3, 2)
a.shape = (3, )
```

同样，规则1告诉我们，a数组的形状必须用1进行补全：

```python
M.shape -> (3, 2)
a.shape -> (1, 3)
```

根据规则2，a数组的第一个维度进行扩展以匹配M的维度：

```python
M.shape -> (3, 2)
a.shape -> (3, 3)
```

现在需要用到规则3——最终的形状还是不匹配，因此这两个数组是不兼容的。当我们执行运算时会看到以下结果：

````python
M + a
````

请注意，这里可能会发生的混淆在于：你可能想通过在a数组的右边补1，而不是左边补1，让a和M的维度变得兼容。但这是不被广播的规则所允许的。这种灵活性在有些情景中可能会有用，但是它可能会导致结果模糊。如果你希望实现右边补全，可以通过变形数组来实现。可以使用np.newaxis关键字。

````python
a[:, np.newaxis].shape
````

````python
M + a[:, np.newaxis]
````

注意，这里仅仅用到了+运算符，而这些广播规则对于任意二进制通用函数都是适用的。例如这里的`logaddexp(a, b)`函数，比起简单的方法，该函数计算`log(exp(a) + exp(b))`更准确：

````python
np.logaddexp(M, a[:, np.newaxis])
````

#### 布尔运算符

我们已经看到该如何统计所有降水量小于4英寸或者大于2英寸的天数，但是如果我们想统计降水量小于4英寸且大于2英寸的天数，应该如何操作呢？这个可以通过python的**逐位逻辑运算符(bitwise logic operator)**&、|、^和~来实现。同标准的算术运算符一样，NumPy用通用函数重载了这些逻辑运算符，这样可以实现数组的逐位运算（通常是布尔运算）。

例如，可以写如下的复合表达式：

````python
np.sum((inches > 0.5) & (inches < 1))
````

可以看到，降水量在0.5英寸~1英寸间的天数是29天。注意这些括号十分重要，因为优先级问题，如果去掉这些括号，该表达式会变成以下形式，这会导致错误：

$$
inches > (0.5 & inches) < 1
$$

利用A AND B 和 NOT(A OR B)的等价原理，可以用另一种形式实现同样的结果：

````python
np.sum(~((inches <= 0.5) | (inches >= 1)))
````

将比较运算符和布尔运算符合并起来用在数组上，可以实现更多有效的逻辑运算操作。以下表格总结了逐位的布尔运算符和其对应的通用函数。

|运算符|对应通用函数|
|:---|:---|
|\&|np.bitwise_and|
|\||np.bitwise_or|
|\^|np.bitwise_xor|
|\~|np.bitwise_not|

利用这些工具，就可以回答之前关于天气数据的问题了。以下的示例是结合使用掩码和聚合实现的结果计算：

````python
print("Number days without rain:       ", np.sum(inches == 0))
print("Number days with ranin:         ", np.sum(inches != 0))
print("Days with more than 0.5 inches: ", np.sum(inches > 0.5))
print("Rainy days with < 0.1 inches:   ", np.sum((inches > 0) & (inches < 0.2)))
````

## 花哨的索引（fancy indexing）

花哨的索引和前面那些简单的索引非常类似，但是传递的是索引数组，而不是单个标量。花哨的索引让我们能够快速获得并修改复杂的数组值得子数据集。

### 探索花哨得索引

花哨的索引意味着：传递一个索引数组来一次性获得多个数组元素。例如以下数组：

````python
import numpy as np
rand = np.random.RandomState(42)

x = rand.randint(100, size=10)
print(x)
````

假设我们希望获得三个不同的元素，可以用以下方式实现：

````python
[x[3], x[7], x[2]]
````

另一种方法就是通过传递索引的单个列表或者数组来获得同样的结果：

````python
ind = [3, 7, 4]
x[ind]
````

利用花哨的索引，结果的形状和索引数组的形状一致，而不是与被索引数组的形状一致：

````python
ind = np.array([[3, 7],
                [4, 5]])
x[ind]
````

花哨的索引也对多个维度适用。假设我们有以下数组：

````python
X = np.arange(12).reshape((3, 4))
X
````

和标准的索引方式一样，第一个索引指的是行，第二个索引指的是列：

````python
row = np.array([0, 1, 2])
col = np.array([2, 1, 3])
X[row, col]
````

这里需要注意，结果的第一个值是X\[0, 2\]，第二个值是X\[1, 1\]，第三个值是X\[2, 3\]。索引值的配对遵循广播的规则。

````python
X[row[:, np.newaxis], col]
````

这里，每一行的值都与每一列的向量配对，正如我们看到的广播的算术运算：

````python
row[:, np.newaxis] * col
````

这里要特别记住，花哨的索引返回的值反映的是广播后索引数组的形状，而不是被索引数组的形状。

### 组合索引

花哨的索引可以和其他索引方案结合起来形成更加强大的索引操作：

````python
print(X)
````

可以将花哨的索引和简单的索引组合使用：

````python
X[2, [2, 0, 1]]
````

也可以将花哨的索引和切片组合使用：

````python
X[1:, [2, 0, 1]]
````

更可以将花哨的索引和掩码组合使用：

````python
mask = np.array([1, 0, 1, 0], dtype=bool)
X[row[:, np.newaxis], mask]
````

### 用花哨的索引修改值

正如花哨的索引可以被用于获取部分数组，它也可以被用于修改部分数组。例如，假设我们有一个索引数组，并且希望设置数组中对应的值：

````python
x = np.arange(10)
i = np.array([2, 1, 8, 4])
x[i] = 99
print(x)
````

可以用于任何的赋值操作来实现，例如：

````python
x[i] -= 10
print(x)
````

不过需要注意，操作中重复的索引回导致一些出乎意料的结果，比如：

````python
x = np.zeros(10)
x[[0, 0]] = [4, 6]
print(x)
````

4到哪里去了？这个操作首先赋值x\[0\] = 4，然后赋值x\[0\] = 6，因此当然值为6，合理。但假如：

````python
i = [2, 3, 3, 4, 4, 4]
x[i] += 1
x
````

你可能期望x\[3\] = 2，x\[4\] = 3，因为这是这些索引重复的次数。但是为什么结果不同于我们的预想呢？从概念上讲，应该是我们期待的结果，然而事实并非如此。

如果你希望累加，应该怎么做？可以使用通用函数中的`at()`方法来实现。

````python
x = np.zeros(10)
np.add.at(x, i, 1)
print(x)
````

`at()`函数在这里对给定的操作、给定的索引（这里是i）以及给定的值（这里是1）执行的是就地操作。另一个可以实现该功能的类似方法是通用函数中的`reduceat()`函数，你可以在`NumPy`文档中找到关于该函数的更多信息。

### 示例：数据区间划分

你可以用这些方法有效地将数据进行区间划分并手动创建直方图。例如，假定我们有1000个值，希望快速统计分布在每个区间中的数据频次，可以用`ufunc.at`来计算：

````python
np.random.seed(42)
x = np.random.randn(100)

# 手动计算直方图
bins = np.linspace(-5, 5, 20)
counts = np.zeros_like(bins)

# 为每个x找到合适的区间
i = np.searchsorted(bins, x)

# 为每个区间加上1
np.add.at(counts, i, 1)
````

计数数组counts反映的是每个区间中的点的个数，即直方图分布：

````python
# 画出结果
plt.plot(bins, counts, linestyle='steps');
````

当然，如果每次需要画直方图你都这么做的话，也是很不明智的。这就是为什么Matplotlib提供plt.hist()方法，该方法仅用一行代码就实现了上述功能：

````python
plt.hist(x, bins, histtype='step');
````

这个函数将生成一个一摸一样的图，为了计算区间，Matplotlib将使用np.histogram函数，该函数的计算功能也和上面执行的计算类似。接下来比较一下这两种方法：

````python
print("NumPy routine:")
%timeit counts, edges = np.histogram(x, bins)

print("Custom routine:")
%timeit np.add.at(counts, np.searchsorted(bins, x), 1)
````

可以看到，我们一行代码的算法比NumPy优化过的算法快好几倍！这是因为NumPy的算法要顾及更多的情况，它在数据点比较大的时候更能显示其良好性能。

````python
x = np.random.randn(1000000)
print("NumPy routine:")
%timeit counts, edges = np.histogram(x, bins)

print("Custom routine:")
%timeit np.add.at(counts, np.searchsorted(bins, x), 1)
````

在将Python有效地用于数据密集型应用中的关键是，当应用场景合适时知道使用np.histogram这样的现成函数，当需要执行更多指定的操作时也知道如何利用更低级的功能来实现。

## 数组的排序

到目前为止，我们主要关注了用NumPy获取和操作数组的工具。这一节主要介绍用于排序NumPy数组的相关算法。

例如，一个简单的选择排序重复寻找列表中的最小值，并且不断交换直到列表是有序的。可以在Python中仅用几行代码来实现：

````python
import numpy as np

def selection_sort(x):
    for i in range(len(x)):
        swap = i + np.argmin(x[i:])
        (x[i], x[swap]) = (x[swap], x[i])
    return x
````

````python
x = np.array([2, 1, 4, 3, 5])
selection_sort(x)
````

但是选择排序太慢了！对于一个包含N个值的数组来说，它需要做N个循环，每个循环中还要执行~N次比较，以找到交换值。选择排序的平均算法复杂度为`O[N^2]`：如果你将列表中元素的个数翻倍，那么运行时间就会延长4倍。

即便如此，选择排序都比我自己最喜欢的bogosort排序算法的性能要好：

````python
def bogosort(x):
    while np.any(x[:-1] > x[1:]):
        np.random.shuffle(x)
    return x
````

````python
x = np.array([2, 1, 4, 3, 5])
bogosort(x)
````

````python
x[:4] > x[1:]
````

这个憨憨的算法完全是碰运气：它不断对数组元素进行随机重排，直到成为有序数组才停止。这个算法的复杂度是$O[N \times N]$。

### NumPy中的快速排序：np.sort和np.argsort

尽管python有内置的sort和sorted函数可以对列表排序，但是这里不介绍这两个函数。因为NumPy的np.sort函数实际上效率更高。默认情况下，np.sort的排序算法是快速排序，其算法复杂度为$O[N \log{N}]$，另外也可以选择归并排序和堆排序。

如果想在不修改原始输入数组的基础上返回一个排序好的数组，可以使用np.sort：

````python
x = np.array([2, 1, 4, 3, 5])
np.sort(x)
````

如果希望用排序好的数组代替原始数组，可以使用数组的sort方法：

````python
x.sort()
print(x)
````

另一个相关的函数时argsort，该函数返回的是原始数组排序好的索引值：

````python
x = np.array([2, 1, 4, 3, 5])
i = np.argsort(x)
print(i)
````

以上结果的第一个元素是数组最小元素的索引值，第二个值给出的是次小元素的索引值，以此类推。这些索引值可以被用于（通过花哨的索引）创建有序的数组：

````python
x[i]
````

### 部分排序：分隔

有时候我们不希望对整个数组进行排序，仅仅希望找到数组中第K小的值，NumPy的np.partition函数提供了该功能。np.partition函数的输入是数组和数字K，输出结果是一个新的数组，最左边是前K小的值，往右是任意顺序的其他值：

````python
x = np.array([7, 2, 3, 1, 6, 5, 4])
np.partition(x, 3)
````

请注意，结果数组中前三个值是数组中最小的三个值，剩下的位置是原始数组剩下的值。在这两个分隔区中，元素都是任意排列的。

还有一个np.argpartition函数计算的是分隔的索引值。

### 示例：K个最近邻

以下示例展示的是如何利用argsort函数沿着多个轴快速找到集合中每个点的最近邻。首先，在二维平面上创建一个有10个随机点的集合。按照惯例，将这些数据点放在一个$10 \times 2$的数组中：

````python
X = rand.rand(10, 2)
````

为了对这些点有一个直观的印象，来画出它的散点图：

````python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

plt.scatter(X[:, 0], X[:, 1], s=100);
````

两点之间的距离的平方是每个维度的距离差的平方和。

$$
d^2 = \sqrt{\sum_{i = 1}^{n}{(x_{2i}-x_{1i})^2}}
$$

利用NumPy的广播和聚合功能，可以用一行代码计算矩阵的平方距离：

````python
dist_sq = np.sum((X[:, np.newaxis, :] - X[np.newaxis, :, :]) ** 2, axis=-1)
dist_sq
````

这个操作看起来有点困惑，因为你还不熟悉NumPy的数组切片和广播规则！你应该将其各组件分解后再分析！

````python
# 在坐标系中计算每对点的差值
differences = X[:, np.newaxis, :] - X[np.newaxis, :, :]
differences.shape
````

> np.newaxis其实是none的别名，用?就能查看到。也许你会对用np.newaxis来变形有不理解的地方，但是，我学会了一个方法，能方便理解它。

````python
# 建立一个 10 * 2 的矩阵
jz = np.random.randint(0, 10, (10, 2))

# 打印它的shape以及具体结构
print(jz)
print(jz.shape)

# 按照方式1变形，并打印出它的shape
print(jz[:, np.newaxis, :].shape)

# 按照方式2变形，并打印出它的shape
print(jz[np.newaxis, :, :].shape)
````

> 可以看到，np.newaxis就是给前一层的元素打多一层的包装。而这个前一层指的是2，我们应该从右往左读(10, 1, 2)：2个元素组成一个组，1个组组成一个更大的组，10个更大的组组成一个更更大的组。

这里要解释一下`X[:, np.newaxis, :] - X[np.newaxis, :, :]`，如果你用笔操作一下，你就会发现它是每个点和其他点距离差，一共有10个点。

````python
# 求出差值的平方
sq_differences = differences ** 2
sq_differences.shape
````

````python
# 将差值求和获得平方距离
dist_sq = sq_differences.sum(-1)
dist_sq.shape
````

这里的axis=-1可能不好理解，应该分成2步来理解：
1. -1是倒数第一个轴的意思
1. axis=k是说第k个轴被折叠的意思，也就是该轴消失，该轴的元素形式不变，但元素个体要以某种方式变成一个个体（代表原先的轴）

请再次确认以上步骤，应该看到该矩阵的对角线都是0：

````python
dist_sq.diagonal()
````

结果的确是这样！！！当我们有了这样一个转化为两点间的平方距离的矩阵后，就可以使用np.argsort函数沿着每行进行排序了。最左边给出的列给出的索引值就是最进邻：

````python
nearest = np.argsort(dist_sq, axis=1)
print(nearest)
print(nearest.shape)
````

需要注意的是，第一列是按0~9从小到大排列的，这是因为每个点的最近邻是它本身。

如果使用全排列，我们实际上可以实现的比这个例子展示的还要多。如果我们仅仅关心k个最近邻，那么唯一要做的是分隔每一行，这样最小的k+1的平方距离将排在最前面，其他更长的距离占据矩阵该行的其他位置。可以用np.argpartition函数实现。

````python
k = 2
nearest_partition = np.argpartition(dist_sq, k + 1, axis=1)
````

为了将邻接点网络可视化，我们将每个点和其最近的两个最近邻连接：

````python
plt.scatter(X[:, 0], X[:, 1], s=100)

# 将每个点与它的两个最近邻连接
K = 2

for i in range(X.shape[0]):
    for j in nearest_partition[i, :K+1]:
        # 画一条从 X[i] 到 X[j] 的线段
        # 用 zip 方法实现：
        plt.plot(*zip(X[j], X[i]), color='black')
````

你可能会看到点的连线多于两条，原因很简单，这里不记录。

广播和按行排序可能看起来不如循环直观，但是更高效。而且向量化操作对输入数据的数据量并不敏感。

最后还想提醒一点，做大数据量的最近邻搜索时，有几种基于树的算法的算法复杂度可以实现接近$O[N \log{N}]$，其中一种就是KD-Tree。

## 结构化数据：NumPy的结构化数组

本节介绍NumPy的 **结构化数组** 和 **记录数组**，它们为复合的、异构的数据提供了非常有效的存储。尽管这里列举的模式对于简单的操作非常有用，但是这些场景通常也可以用Pandas的DataFrame来实现（将在第三章详细介绍）。

如果类型的名称对你来说不重要，那你可以仅仅用一个字符串来指定它。在该字符串中数据类型用逗号分隔：

````python
np.dtype('S10, i4, f8')
````

简单的字符串格式的代码可能看起来令人困惑，但是它们其实基于非常简单的规则。第一个（可选）字符是<或者>，分别表示“低字节序”（little endian）和“高字节序”（bidendian），表示字节（bytes）类型的数据在内存中存放顺序的习惯用法。后一个字符指定的是数据的类型：字符、字节、整型、浮点型，等等。最后一个字符表示该对象的字节大小。

### 更高级的复合类型

NumPy中也可以定义更高级的复合数据类型。例如，你可以创建一种类型，其中每个元素都包含一个数组或矩阵。我们会创建一个数据类型，该数据类型用mat组件包含一个$3 \times 3$的浮点矩阵：

````python
tp = np.dtype([('id', 'i8'), ('mat', 'f8', (3, 3))])
X = np.zeros(1, dtype=tp)
print(X[0])
print(X['mat'][0])
````

> 记住：括号是元组，一种不可变的序列。

现在X数组的每个元素都包含一个id和一个$3 \times 3$的矩阵，为什么我们宁愿用这种方法存储数据，也不愿意用简单的多维数组或者python字典呢？

因为NumPy的dtype直接映射到C结构的定义，因此包含数组内容的缓存可以直接在C程序中使用。如果你想写一个Python接口与一个遗留的C语言或者Fortran库交互，从而操作结构化数据，你将发现结构化数组非常有用！

### 记录数组：结构化数组的扭转

python代码的设计原则之一是“显式优于隐式”。使用loc和iloc可以让代码更容易维护，可读性更高。特别是在处理整数索引的对象时，我们强烈建议使用这两种索引器。它们既可以让代码阅读和理解起来更加容易，也能避免因误用索引/切片而产生的小bug。

## Pandas数值运算方法

NumPy的基本能力之一就是快速对每个元素进行运算，既包括基本算术运算（加减乘除），也包括更复杂的运算（三角函数、指数函数和对数函数等）。Pandas继承了NumPy的功能，通用函数是关键。

但是Pandas还有其他高级功能：
1. 对于一元运算（像函数和三角函数），这些通用函数将在输出结果中**保留索引**和**列标签**；
1. 而对于二元运算（像加法和乘法），Pandas在传递通用函数时会自动**对齐索引**进行计算；

这就意味着，保存数据内容和组合不同来源的数据——两处在NumPy中都容易出错的地方——变成了Pandas的杀手锏。后面还会介绍一些关于一维Series和二维DataFrame的便捷运算方法。

### 通用函数：保留索引

## 处理缺失值

大多数教程里使用的数据与现实工作中的数据的区别在于后者很少是干净整洁的，许多目前流行的数据集都会有数据缺失的现象。另外，处理不同数据源缺失值的方法还不同。

本节介绍一些处理缺失值的通用规则，Pandas对缺失值的表现形式，并演示Pandas自带的几个处理缺失值的工具的使用方法。本节以及全书涉及的缺失值主要有3种形式：
1. null
1. NaN
1. NA

### 选择处理缺失值的方法

在数据表或者DataFrame中有很多识别缺失值的方法，一般情况下可以分为2种：
1. 通过覆盖全局的掩码表示缺失值
1. 用一个**标签值**（sentinel value）表示缺失值

在掩码的方法中，掩码可能是一个与原数组维度相同的完整布尔类型数组，也可能是用一个比特（0或者1）表示有缺失值的局部状态。

在标签方法中，标签可能是具体的数据（例如用-9999表示缺失的整数），也可能是些极少出现的形式。另外，标签值还可能是更全局的值，比如用一个NaN表示缺失的浮点数，它是IEEE浮点数规范中指定的特殊字符。

使用这两种方法之前都需要先综合考量：使用单独的掩码数组会额外出现一个布尔类型数组，从而增加存储与计算的负担；而标签方法缩小了可以被表示为有效值的范围看，可能需要在CPU或GPU算术逻辑单元中增加额外的（往往也不是最优的）计算逻辑。通用的NaN也不能表示所有数据类型。

大多数情况下，都不存在最佳选择，不同的编程语言与系统使用不同的方法。例如，R语言在每种数据类型中保留一个比特作为缺失值的标签值，而SciDB系统会在每个单元后面加一个额外的字节表示NA状态。

### Pandas的缺失值

Pandas使用了NumPy包的方式处理缺失值，并没有为浮点数据类型提供内置的NA作为缺失值。

NumPy支持掩码数据，也就是说可以用一个布尔掩码数组为原数组标注“无缺失值”或“有缺失值”。Pandas也集成了这个功能，但是在存储、计算和编码维护方面都需要耗费不必要的资源，因此不采用。

Pandas最后选择标签方法表示缺失值，包括两种Python原有的缺失值：

1. 浮点数类型的NaN值
1. Python的None对象

实际效果还不错。

#### None：Python对象类型的缺失值

Pandas可以使用的第一种缺失值标签是None，它是一个Python单体对象，经常在代码中表示缺失值。由于None是一个Python对象，所以不能作为任何NumPy/Pandas数组类型的缺失值，只能用于'object'数组类型（即由Python对象构成的数组）：

````python
import numpy as np
import pandas as pd
````

````python
vals1 = np.array([1, None, 3, 4])
vals1
````

这里dtype=object表示NumPy认为由于这个数组是Python对象构成的，因此将其类型判断为object。虽然这种类型在某些情景中非常有用，对数据的任何操作最终都会在Python层面完成，但是在进行常见的快速操作时，这种类型比其他原生类型要消耗更多的资源：

````python
for dtype in ['object', 'int']:
    print("dtype=", dtype)
    %timeit np.arange(1E6, dtype=dtype).sum()
    print()
````

````python
type1 = np.arange(10, dtype=int).dtype
type2 = np.arange(10, dtype='int').dtype
type1 is type2
````

使用Python对象构成的数组就意味着如果你对一个包含None的数据进行累计操作，如sum()或者min()，那么通常会出现类型错误：

````python
vals1.sum()
````

这就是说，在python中没有定义整数和None之间的加法运算。

#### NaN：数值类型的缺失值

另一种缺失值的标签是NaN（全称Not a Number），是一种按照IEEE浮点数标准设计、在任何系统中都兼容的特殊浮点数：

````python
vals2 = np.array([1, np.nan, 3, 4])
vals2.dtype
````

请注意，NumPy会为这个数组选择一个原生浮点类型，这意味着和之前的object类型数据不同，这个数组会被编译成C代码从而实现快速操作。你可以把NaN看作一个数据类型病毒——它会将它接触过的数据同化。无论和NaN进行什么操作，最终结果都是NaN：

````python
1 + np.nan
````

````python
0 * np.nan
````

虽然这些累计操作的结果定义是合理的（即不会抛出异常），但是并非总是有效的：

````python
vals2.sum(), vals2.min(), vals2.max()
````

NumPy也提供了一些特殊的累计函数，它们可以忽略缺失值的影响：

````python
np.nansum(vals2), np.nanmin(vals2), np.nanmax(vals2)
````

谨记，NaN是一种特殊的浮点数，不是整数、字符串以及其他数据类型。

#### Pandas中NaN与None的差异

虽然NaN和None各有各的用处，但是Pandas把它们视为等价可换的：

````python
pd.Series([1, np.nan, 2, None])
````

Pandas会将没有标签值的数据类型自动转换为NA。例如，当我们将整型数组中的一个值设置为np.nan时，这个值就会强制转换为浮点数缺失值NA。

````python
x = pd.Series(range(2), dtype=int)
x
````

````python
x[0] = None
x
````

请注意，除了整型数组的缺失值强制转换为浮点数，Pandas还会自动将None转换为NaN。




需要注意的是，Pandas 中字符串类型的数据通常是用 object 类型存储的。

### 处理缺失值

Pandas基本上把None和NaN看成一样的。为了完成这个交换过程，Pandas提供了一些方法来发现、剔除、替换数据结构中的缺失值，主要包括以下几种。
1. `isnull`：创建一个布尔类型的掩码标签缺失值。
1. `notnull`：与`isnull()`操作相反。
1. `dropna()`：返回一个剔除缺失值的数据。
1. `fillna()`：放回一个填充了缺失值的数据副本。

#### 发现缺失值

Pandas数据结构有2种有效的方法可以发现缺失值：`isnull()`和`notnull()`。每种方法都是返回布尔类型的掩码数据，例如：

````python
data = pd.Series([1, np.nan, 'hello', None])
````

````python
data.isnull()
````

````python
data.notnull()
````

就好像3.3节介绍的那样，布尔类型掩码数组还可以直接作为Series或者DataFrame的索引使用：

````python
data[data.notnull()]
````

在Series里使用的isnull()和notnull()同样适合DataFrame，产生的结果同样是布尔类型。

#### 剔除缺失值

除了前面介绍的掩码方法，还有两种很好用的缺失值处理方法，分别是dropna()（剔除缺失值）和fillna()（填充缺失值）。在Series上使用这些方法非常简单：

````python
data.dropna()
````

而在DataFrame上使用它们时需要设置一些参数，例如下面的DataFrame：

````python
df = pd.DataFrame([[1,      np.nan, 2],
                   [2,      3,      5],
                   [np.nan, 4,      6]])
df
````

我们没法从DataFrame中单独剔除一个值，要么是剔除缺失值所在的整行，要么是整列。根据实际需要，有时你需要剔除整行，要么是整列，DataFrame中的dropna()会有一些参数可以配置。

默认情况下，dropna()会剔除任何包含缺失值的整行数据：

````python
df.dropna()
````

可以设置按不同的坐标轴剔除缺失值，比如axis=1（或axis='columns'）会剔除任何包含缺失值的整列数据：

````python
df.dropna(axis='columns')
````

但是这样做会把非缺失值一并删除，而有时候只需要剔除全部是缺失值的行或列，或者绝大多数是缺失值的行或列。这些需求可以通过设置how或者thresh参数来满足，它们可以设置剔除行或列缺失值的数量阈值。

默认设置是how='any'，也就是说只要有缺失值就剔除整行或者整列（通过axis设置坐标轴）。你还可以设置how='all'，这样就只会剔除全部是缺失值的行或者列了：

````python
df[3] = np.nan
df
````

````python
df.dropna(axis=1, how='all')
````

还可以通过thresh参数设置行或列中**非缺失值**的最小数量，从而实现更加个性化的配置：

````python
df.dropna(axis='rows', thresh=3)
````

第1行和第3行被剔除了，因为它们只包含2个非缺失值。

#### 填充缺失值

有时候你可能并不想移除缺失值，而是想把它们替换成有效的数值。有效的数值可能是像0、1、2那样单独的值，也可能是经过填充(imputation)或者转换(interpolation)得到的。虽然你可以通过isnull()方法建立掩码来填充缺失值，但是Pandas为此专门提供了一个fillna()方法，它将返回填充了缺失值后的数组副本。

下面用Series来演示：

````python
data = pd.Series([1, np.nan, 2, None, 3], index=list('abcde'))
data
````

我们用一个单独的值来填充缺失值，例如用0：

````python
data.fillna(0)
````

可以用缺失值前面的有效值来从前往后填充(forward-fill)：

````python
data.fillna(method='ffill')
````

也可以用缺失值后面的有效值来从后往前填充(back-fill)：

````python
data.fillna(method='bfill')
````

DataFrame的操作方法与Series类似，只是在填充的时候需要设置坐标轴参数axis：

````python
df
````

````python
df.fillna(method='ffill', axis=1)
````

需要注意的是，假如在从前往后填充时，需要填充的缺失值前面没有值，那么它就仍然是缺失值。

## 层级索引

当目前为止，我们接触的都是一维和二维的数据，但还会有多维数据的需求，它们的索引超过一两个键。因此，Pandas提供了Panel和Panel4D对象解决三维和四维数据。而在实践中，更直观的形式是通过**层级索引**，也被称为**多级索引**配合多个有不同等级(**level**)的一级索引一起使用，这样就可以将高维数组转换成类似一维Series和二维DataFrame对象的形式。

这一节我们介绍创建MultiIndex对象的方法，多级索引数据的取值、切片和统计值的计算，以及普通索引与层级索引的转换方法：

首先导入必要的包：

````python
import pandas as pd
import numpy as np
````

### 多级索引的Series

让我们看看怎么用一维的Series对象表示二维数据——用一系列包含特征与数值的数据点来简单演示。

#### 笨方法

假设你想要分析美国各州在两个不同年份的数据。如果你用前面介绍的Pandas工具来处理，那么可能会用一个Python元组来表示索引。

````python
index = [('California', 2000), ('California', 2010),
         ('New York', 2000), ('New York', 2010),
         ('Texas', 2000), ('Texas', 2010)]
populations = [33871648, 37253956, 18976457, 19378102, 20851820, 25145561]
pop = pd.Series(populations, index=index)
pop
````

通过元组构成的多级索引，你可以直接在Series上取值或用切片查询数据：

````python
pop[('California', 2010):('Texas', 2000)]
````

但是这样做很不方便，假如你想要选择所有2000年的数据，那么就要用一些比较复杂的（可能也比较慢的）清理方法了：

````python
pop[[i for i in pop.index if i[1] == 2010]]
````

这样做不够高效，也不方便。

#### 好办法：Pandas多级索引

Pandas提供了更好的方法，用元组表示索引其实是多级索引的基础，Pandas的MultiIndex类型提供了更丰富的操作方法。我们可以用元组创建一个多级索引，如下所示：

````python
index = pd.MultiIndex.from_tuples(index)
index
````

> 上面和书本说的不太一样。为什么？

如果将前面创建的pop索引重置(reindex)为MultiIndex，就会看到层级索引：

````python
pop = pop.reindex(index)
pop
````

其中前两列表示Series的多级索引值，第三列是数据。你会发现有些行仿佛缺失了第一列数据——这其实是多级索引的表现形式，每个空格与上面的索引相同。

现在可以直接用第二个索引获取2010年的全部数据，与Pandas的切片查询用法一致：

````python
pop[:, 2010]
````

结果是单索引的数组，正是我需要的。与之前的元组索引相比，多级索引的语法更加简单。下面继续介绍层级索引的取值操作方法。

#### 高级数据的多级索引

你可能已经注意到了，我们其实完全可以用一个带行列索引的简单DataFrame代替前面的多级索引。其实Pandas已经实现了类似的功能。unstack()方法可以快速将一个多级索引的Series转化成普通索引的DataFrame：

````python
pop_df = pop.unstack()
pop_df
````

当然，也有stack()方法实现的相反的效果：

````python
pop_df.stack()
````

我们为什么要研究层级索引？因为如果我们可以用含多级索引的一级Series数据表示二维数据，那么我们就可以用Series或DataFrame表示三维甚至更高维度的数据。多级索引每增加一级，就表示数据增加一维，利用这个特点就可以轻松地表示任意维度的数据了。假如要增加一列显示每一年各州的人口统计指标（例如18周岁以下的人口），那么对于这种带有MultiIndex的对象，增加一列就像DataFrame的操作一样简单：

````python
pop_df = pd.DataFrame({'total':pop,
                       'under18':[9267089, 9284094,
                                4687374, 4318033,
                                5906301, 6879014]})
pop_df
````

另外，所以在3.4节介绍过的通用函数和其他功能也同样适用于层级索引。我可以计算上面数据中18岁以下的人口占总人口的比例：

````python
f_u18 = pop_df['under18'] / pop_df['total']
f_u18.unstack()
````

同样，我们可以快速浏览和操作高维数据。

### 多级索引的创建方法

为Series或DataFrame创建多级索引最直接的办法就是将index参数设置为至少二维的索引数组，如下所示：

````python
df = pd.DataFrame(np.random.rand(4, 2), 
                  index = [['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                  columns = ['data1', 'data2'])
df
````

MultiIndex的创建工作将在后台完成。

同理，如果你把将元组作为键的字典传递给Pandas，Pandas也会默认转换为MultiIndex：

````python
data = {('California', 2000):33871648, 
        ('California', 2010):37253956,
        ('Texas', 2000):20851820,
        ('Texas', 2010):25145561,
        ('New York', 2000):18976457,
        ('New York', 2010):19378102}
pd.Series(data)
````

但是有时候显式地创建MultiIndex也是很有用，下面来介绍一些创建方法。

#### 显式地创建多级索引

你可以用pd.MultiIndex中的类方法更加灵活地构建多级索引。例如，就像前面介绍的，你可以通过一个有不同等级的若干简单数组组成的列表来构建MultiIndex：

````python
pd.MultiIndex.from_arrays([['a', 'a', 'b', 'b'], [1, 2, 1, 2]])
````

还可以用两个索引的笛卡尔积创建MultiIndex：

````python
import pandas as pd
import numpy as np
pd.MultiIndex.from_product([['a', 'b'], [1, 2]])
````

更可以直接提供levels（包含每个等级的索引值列表的列表）和labels（包含每个索引值标签列表的列表）创建MultiIndex：

````python
pd.MultiIndex(levels=[['a', 'b'], [1, 2]], 
              labels=[[0, 0, 1, 1], [0, 1, 0, 1]])
````

在创建Series或者DataFrame时，可以将这些对象作为index参数，或者通过reindex方法更新Series或者DataFrame的索引。

#### 多级索引的等级名称

给MultiIndex的等级加上名称会为一些操作提供便利。你可以在前面任何一个MultiIndex构造器中通过names参数设置等级名称，也可以在创建之后通过索引的names属性来修改名称：

````python
pop.index.names = ['state', 'year']
pop
````

在处理复杂的数据时，为等级设置名称是管理多个索引值的好办法。

## 合并数据集：Concat与Append操作

将不同的数据源进行合并是数据科学中最有趣的事情之一，这既包括将2个不同的数据集非常简单地拼接在一起，也包括用数据库那样的连接(join)和合并(merge)操作处理有重叠字段的数据集。Series与DataFrame都具备这类操作，Pandas的函数与方法让数据合并变得快速简单。

先来用pd.concat函数演示一个Series与DataFrame的简单合并操作。之后，我们将介绍Pandas中更复杂的merge和join内存数据合并操作。

首先导入Pandas和NumPy：

````python
import pandas as pd
import numpy as np
````

简单起见，定义一个能创建DataFrame某种形式的函数，后面将会用到：

````python
def make_df(cols, ind):
    """一个简单的DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)

# 示例
make_df('ABC', range(3))
````

### 知识回顾：NumPy数组的合并

合并Series与DataFrame与合并NumPy数组基本相同，后者通过2.2节介绍的np.concatenate函数即可完成。你可以用这个函数将两个或者两个以上的数组合并成一个数组。

````python
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

np.concatenate([x, y, z])
````

第一个参数是需要合并的数组列表或者元素，还有一个axis参数可以设置合并的坐标轴方向：

````python
x = [[1, 2], [3, 4]]
np.concatenate([x, x], axis=1)
````

### 通过pd.concat实现简易合并

Pandas有一个pd.concat()函数与np.concatenate语法类似，但是配置参数更多，功能更强大：

```python
# Pandas 0.18版中的函数签名
pd.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
```

pd.concat()可以简单地合并一维的Series或DataFrame对象，与np.concatenate()合并数组一样：

````python
ser1 = pd.Series(['A', 'B', 'C'], index=[1, 2, 3])
ser2 = pd.Series(['D', 'E', 'F'], index=[4, 5, 6])
pd.concat([ser1, ser2])
````

它也可以用来合并高维数据，例如下面的DataFrame：

````python
df1 = make_df('AB', [1, 2])
df2 = make_df('AB', [3, 4])
print(df1); print(df2); print(pd.concat([df1, df2]))
````

默认情况下，DataFrame的合并都是逐行进行的（默认设置是axis=0）。与np.concatenate()一样，pd.concat()也可以设置合并坐标轴，例如下面的的示例：

````python
df3 = make_df('AB', [0, 1])
df4 = make_df('CD', [0, 1])
print(df3);print(df4);print(pd.concat([df3, df4], axis='columns'))
````

#### 索引重复

np.concatenate和pd.concat最主要的差异之一就是Pandas在合并时会**保留索引**，即使索引是重复的！例如：

````python
x = make_df('AB', [0, 1])
y = make_df('AB', [2, 3])
y.index = x.index    # 复制索引
print(x);print(y);print(pd.concat([x, y]))
````

你会发现结果中的索引是重复的。虽然DataFrame允许这么做，但结果并不是我想要的。pd.concat()提供了一些解决这个问题的方法。

##### （1）捕捉索引重复的错误

如果你要检测pd.concat()合并的结果是否出现了重复的索引，可以设置verify_integrity参数。将参数设置为True，合并时若有索引重复就会触发异常。下面的示例可以让我们清晰地捕捉并打印错误信息： 

````python
try:
    pd.concat([x, y], verify_integrity=True)
except ValueError as e:
    print("ValueError: ", e)
````

##### （2）忽略索引

有时候索引无关要紧，那么合并时就可以忽略它们，可以通过设置ignore_index参数来实现。如果将参数设置为True，那么合并时会创建一个新的整数索引。

````python
print(x);print(y);print(pd.concat([x, y], ignore_index=True))
````

##### （3）增加多级索引

另一种处理索引重复的方法是通过keys参数为数据源设置多级索引标签，这样结果就会带上多级索引：

````python
print(x);print(y);print(pd.concat([x, y], keys=['x', 'y']))
````

示例合并后的结果是多级索引的DataFrame，可以用3.6节中介绍的方法将它们转换成我们需要的形式。

#### 类似join的合并

前面介绍的简单例子都有一个共同点，那就是合并的DataFrame都是同样的列名。而在实际工作中，需要合并的数据往往带有不同的列名，而pd.concat提供了一些选项来解决这类合并问题。看下面两个DataFrame，它们的列名部分相同，却又不完全相同：

````python
df5 = make_df('ABC', [1, 2])
df6 = make_df('BCD', [3, 4])
print(df5);print(df6);print(pd.concat([df5, df6]))
````

默认情况下，某个位置上缺失的数据会用NaN表示。如果不想这样，可以用join和join_axes参数设置合并方式。默认的合并方式是对所有输入列进行并集合并（join='outer'），当然也可以用join='inner'实现对输入列的交集合并：

````python
print(df5);print(df6);
print(pd.concat([df5, df6], join='inner'))
````

另一种合并方式是直接确定结果使用的列名，设置join_axes参数，里面是索引对象构成的列表（是列表的列表）。如下面示例所示，将结果的列名设置为第一个输入的列名：

````python
print(df5);print(df6);
print(pd.concat([df5, df6], join_axes=[df5.columns]))
````

pd.concat的合并功能可以满足你在合并两个数据集时的许多需求，操作时请记住一点。

#### append()方法

因此直接进行数组合并的需求非常普遍，所以Series和DataFrame对象都支持append方法，让你使用最少的代码实现合并功能。例如，你可以使用df1.append(df2)，效果与pd.concat(\[df1, df2\])一样：

````python
print(df1);print(df2);print(df1.append(df2))
````

需要注意的是，与Python列表中的append()和extend()方法不同，Pandas的append()不直接更新原有对象的值，而是为合并的数据创建一个新对象。因此，它不能被称为一个很高效的解决方案，因为每次合并都需要重新创建索引和数据缓存。总之，如果你需要进行多个append操作，还是建议先创建一个DataFrame列表，然后用concat()函数一次性解决所有合并任务。

下一节介绍另一种功能强大的数据组合方法——类似数据库的数据合并，在pd.merge里实现。关于concat()和append()更多的信息，请参考Pandas文档中“Merge, Join, and Concatenate”节。

## 合并数据集：合并与连接

Pandas的基本特性之一就是高性能的内存式数据连接(join)与合并(merge)操作。如果你有使用数据库的经验，那么对这类操作一定很熟悉。Pandas的主接口是pd.merge函数，下面让我们通过一些示例来介绍它的用法。

### 关系代数

pd.merge()实现的功能基于**关系代数(relation algebra)**的一部分。关系代数是处理关系型数据库的通用理论，绝大部分数据库的可用操作都以此为理论基础。关系代数方法论的强大之处在于，它提出的若干简单操作规则经过组合就可以为任意数据集构建十分复杂的操作。借助在数据库或程序里已经高效实现的基本操作规则，你可以完成很多复杂的操作。

Pandas在pd.merge()函数和Series和DataFrame的join()方法里实现了这些基本操作规则。下面来看看如何用这些简单的规则连接不同数据源的数据。

### 数据连接的类型

pd.merge()函数实现了3种数据连接的类型：

1. 一对一
1. 多对一
1. 多对多

这三种数据连接类型都通过pd.merge()接口进行调用，需求不同操作不同。

#### 一对一连接

例如，2个包含同一所公司员工不同信息的DataFrame：

````python
df1 = pd.DataFrame({'employee':['Bob', 'Jake', 'Lisa', 'Sue'], 
                    'group':['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee':['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date':[2004, 2008, 2012, 2014]})
print(df1);print(df2)
````

若想将这两个DataFrame合并成一个DataFrame，可以用pd.merge()函数实现：

````python
df3 = pd.merge(df1, df2)
df3
````

pd.merge()方法会发现两个DataFrame都有“employee”列，并会自动以这列作为键进行连接。两个输入的合并结果是一个新的DataFrame。需要注意的是，共同列的位置可以是不一致的。例如在这个例子中，虽然df1和df2中“employee”列的位置是不一样的，但是pd.merge()函数会正确处理这个问题。另外还需要注意的是，pd.merge()会默认丢弃原来的行索引，不过也可以自定义（详情参见3.8.3）。

#### 多对一连接

多对一连接是指，在需要连接的2个列中，有一列的值有重复。通过多对一连接获得的结果DataFrame将会保留重复值。请看下面的例子：

````python
df4 = pd.DataFrame({'group':['Accounting', 'Engineering', 'HR'],
                    'supervisor':['Carly', 'Guido', 'Steve']})
print(df3);print(df4);print(pd.merge(df3, df4))
````

“多对一”意思是相同的那2列的其中一列有重复的元素，而另一列没有，为了能连接成功，另一列相应部分也要重复。

#### 多对多连接

多对多连接就是，相同的那两列都有重复值。用一个例子讲解一下，这里有一个DataFrame显示不同岗位人员的一种或多种能力：

````python
df5 = pd.DataFrame({'group':['Accounting', 'Accounting',
                             'Engineering', 'Engineering', 'HR', 'HR'],
                    'skills':['math', 'spreadsheets', 'coding', 'linux',
                              'spreadsheets', 'organization']})
print(df1);print(df5);print(pd.merge(df1, df5))
````

这三种数据连接类型可以直接与其他Pandas工具组合使用，从而实现各种各样的功能。但是工作中的真实数据集往往不像示例中演示的那么干净、整洁。下面就来介绍pd.merge()的一些功能，它们可以让你更好地处理数据连接中的问题。

### 设置数据合并的键

我们已经见过pd.merge()的默认行为了：它会将两个输入的一个或多个共同列作为键进行合并。但由于输入要合并的列通常都是不同名的，因此pd.merge()提供了一些参数处理这个问题。

#### 参数on的用法

最简单的方法就是直接将参数on设置为一个列名字符串或者一个包含多列名称的列表：

````python
print(df1);print(df2);print(pd.merge(df1, df2, on='employee'))
````

这个参数只能在两个DataFrame有共同列名的时候才可以使用。

#### left_on与right_on参数

有时候要合并两个列名不同的数据集，例如前面的员工信息表中有一个字段不是“employee”而是“name”。可以这么做：

````python
df3 = pd.DataFrame({'name':['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary':[70000, 80000, 120000, 90000]})
print(df1);print(df3);
print(pd.merge(df1, df3, left_on="employee", right_on="name"))
````

获取的结果中会有一个多余的列，可以通过DataFrame的drop()方法划掉：

````python
pd.merge(df1, df3, left_on="employee", right_on="name").drop('name', axis=1)
````

#### left_index与right_index参数

除了合并列之外，你可能还需要合并索引。例如：

````python
df1a = df1.set_index('employee')
df2a = df2.set_index('employee')
print(df1a);print(df2a)
````

你可以通过设置pd.merge()中的left_index和/或right_index参数将索引设置为键来实现合并：

````python
print(df1a);print(df2a);
print(pd.merge(df1a, df2a, left_index=True, right_index=True))
````

为了方便考虑，DataFrame实现了join()方法，它可以按照索引进行数据合并：

````python
print(df1a);print(df2a);print(df1a.join(df2a))
````

如果想将索引和列混合使用，那么可以通过结合left_index和right_on，或者结合left_on与right_index来实现：

````python
print(df1a);print(df3);
print(pd.merge(df1a, df3, left_index=True, right_on='name'))
````

这些参数都适用于多个索引和/或多个列名，函数接口非常简单。若想了解更多，查看Pandas文档。

### 设置数据连接的集合操作规则

通过前面的示例，我们总结出数据连接的一个重要连接：集合操作规则。当一个值出现在一列，却没有出现在另一列时，就需要考虑集合操作规则了。比如：

````python
df6 = pd.DataFrame({'name':['Peter', 'Paul', 'Mary'],
                    'food':['fish', 'beans', 'bread']},
                   columns=['name', 'food'])
df7 = pd.DataFrame({'name':['Mary', 'Joseph'],
                    'drink':['wine', 'beer']},
                   columns=['name', 'drink'])
print(df6);print(df7);print(pd.merge(df6, df7))
````

我们合并两个数据集，在“name”列中只有一个共同的值：Mary。默认情况下，结果中只会包含2个输入集合的**交集**，这种连接方式被称为**内连接**(inner join)。我们可以用how参数设置连接方式，默认值为'inner'：

````python
pd.merge(df6, df7, how='inner')
````

how参数支持的数据连接方式还有'outer', 'left'和'right'。外连接（outer join）返回两个输入列的交集，所有缺失值都用NaN填充：

````python
print(df6);print(df7);print(pd.merge(df6, df7, how='outer'))
````

> 内连接和外连接有什么不同呢？主要的不同在于，内连接只保留交集，外连接会把没有相交的部分也留下来，缺失部分用NaN填充。

左连接和右连接返回的结果分别只包含左列和右列，如下所示：

````python
print(df6);print(df7);print(pd.merge(df6, df7, how='left'))
````

> 这里的意思是说，外连接是把两个表格的不相交部分都留下，但是左连接是只留左边表格的，右连接是只留右边表格的。这里4种连接其实就是靠保留非相交部分的程度不同得以区别的。

### 重复列名：suffixes参数

最后，你可能会遇到两个输入DataFrame有重名列的情况。来看看下面的例子：

````python
df8 = pd.DataFrame({'name':['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank':[1, 2, 3, 4]})
df9 = pd.DataFrame({'name':['Bob', 'Jake', 'Lisa', 'Sue'],
                    'rank':[3, 1, 4, 2]})
print(df8);print(df9);print(pd.merge(df8, df9, on='name'))
````

### Groupby：分割、应用和组合

简单的累计方法可以让我们对数据集有一个笼统的认识，但是我们经常还需要对某些标签或索引的局部进行累计分析，这时就要用到groupby了。虽然“分组(group by)”这个名字是借用SQL数据库语言的命令，但是其理念引用发明R语言frame的Hadley Wickham的观点可能更加合适：
1. 分割(split)
1. 应用(apply)
1. 组合(combine)

#### 分割、应用和组合

一个经典的分割-应用-组合操作示例如下图所示，其中“apply”的是一个求和函数。

步骤如下：
1. 分割步骤：按照指定的键，分割成若干组
1. 应用步骤：对每个组分别应用操作
1. 组合步骤：将每一个组的结果合并成一个输出组



我们需要意识到的是：**这个中间的过程是隐式的。**这个就是Groupby的好处，它把操作封装抽象了，所以用户只要知道它能做什么就好了，不必自己考虑内部实现。

用Pandas进行图3-1所示的计算作为具体的示例。从创建输入DataFrame开始：

````python
df = pd.DataFrame({'key':['A', 'B', 'C', 'A', 'B', 'C'],
                   'data':range(6)}, columns=['key', 'data'])
df
````

使用DataFrame的groupby()方法即可：

````python
df.groupby('key')
````

需要注意的是，这里返回的是一个DataFrameGroupBy对象，这个对象的神奇之处在于，里面隐藏着若干组数据，但是在没有应用累计函数之前不会计算。这种延迟计算的方式使得大多数的累计操作都可以通过一种对用户几乎透明的方式高效执行。

> 透明：我终于理解什么叫做透明了！意思是说，用户看不见！并不是说用户能看得更清楚，我总是想象成玻璃柜，但其实我应该想象成玻璃！

接下来对DataFrameGroupBy对象应用累计函数，它会完成相应的步骤并生成结果：

````python
df.groupby('key').sum()
````

你可以使用Pandas中任意的一种累计函数，也可以用任意有效的DataFrame对象。

#### GroupBy对象

GroupBy对象是一种非常灵活的抽象类型。在大多数场景下，你都可以将其看成DataFrame的集合，在底层解决所有的难题。下面用行星数据做演示。

GroupyBy中最重要的操作可能就是aggregate、filter、transform和apply（累计、过滤、转换）了，后文将详细介绍这些操作，现在先来介绍一些GroupBy的基本操作方法。

（1）按列取值。GroupyBy对象与DataFrame一样，也支持按列取值，并返回一个修改过的GroupBy对象，例如：

````python
planets.groupby('method')
````

````python
planets.groupby('method')['orbital_period']
````

可以看到这里是从原来的DataFrame中取了某个列名作为一个Series组，但和前面一样，直到我们运算累计函数，才会计算：

````python
planets.groupby('method')['orbital_period'].median()
````

（2）按组迭代。GroupBy对象支持直接按组迭代，返回的每一组都是Series或DataFrame：

````python
for (method, group) in planets.groupby('method'):
    print("{0:30s} shape={1}". format(method, group.shape))
````

虽然使用内置的apply功能还要快，但这个方式在处理某些问题的时候十分有用。

（3）调用方法。借助Python的魔力，可以让任何不由GroupBy直接实现的方法直接应用到每一组，无论是DataFrame还是Series对象都是同样适用的。例如，你可以用DataFrame的describe()方法进行累计，对每一组数据进行描述性统计：

````python
planets.groupby('method')['year'].describe().unstack()
````

这就是分割、应用、组合。

#### 累计、过滤、转换和应用

前面的章节仅仅是介绍了组合操作，还有GroupBy对象的aggregate(), filter(), transform()和apply()方法，这些方法在数据组合之前实现了大量高效的操作。

为了方便后面内容的演示，使用下面这个DataFrame：

````python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key':['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1':range(6),
                   'data2':rng.randint(0, 10, 6)},
                  columns = ['key', 'data1', 'data2'])
df
````

（1）累计。除了sum()和median()之类的简单函数，aggregate()能支持更复杂的操作。比如字符串、函数或者函数列表，并且能一次性计算所有的累计值。例如：

````python
df.groupby('key').aggregate(['min', np.median, max])
````

````python
df.groupby('key').aggregate
````

另一种用法就是通过Python字典指定不同列需要累计的函数：

````python
df.groupby('key').aggregate({'data1':'min',
                             'data2':'max'})
````

（2）过滤。过滤操作可以让你按照分组的属性丢弃若干数据。例如，我们可能只需要保留标准差超过某个阈值的组：

````python
def filter_func(x):
    return x['data2'].std() > 4

print(df);print(df.groupby('key').std());
print(df.groupby('key').filter(filter_func))
````

filter()函数会返回一个布尔值，表示每个组是否通过过滤。

（3）**转换**。累计操作返回的是对组内全量数据缩减过的结果，而转换操作会返回一个新的全量数据。数据经过转换之后，其形状与原来的输入数据是一样的。常见的例子就是将每一组的样本数据减去各组的均值，实现数据标准化。

````python
df.groupby('key').transform(lambda x: x - x.mean())
````

（4）apply()方法。apply()方法让你可以在每个组上应用任意方法。这个函数输入一个DataFrame，返回一个Pandas对象（DataFrame或Series）或一个标量（scalar，单个数值）。组合操作会适应返回结果类型。

下面的例子就是用apply()方法将第一列数据以第二列的和为基数进行标准化：

````python
def norm_by_data2(x):
    # x是一个分组数据的DataFrame
    x['data1'] /= x['data2'].sum()
    return x

print(df);print(df.groupby('key').apply(norm_by_data2))
````

GroupBy里的apply()方法非常灵活，唯一需要注意的地方是它总是输入分组数据的DataFrame，返回Pandas对象或标量。具体如何选择需要视情况而定。

#### 设置分割的键

````python
L = [0, 1, 0, 1, 2, 0]
print(df);print(df.groupby(L).sum());
````

````python
df2 = df.set_index('key')
mapping = {'A':'vowel', 'B':'consonant', 'C':'consonant'}
print(df2);print(df2.groupby(mapping).sum())
````

（3）**任意Python函数。**与前面的字典映射类似，你可以将任意Python函数传给groupby，函数映射到索引，然后新的分组输出：

## 数据透视表

我们已经介绍过GroupBy抽象类是如何探索数据集内部的关联性了，**数据透视表(pivot table)** 是一种类似的操作方法，常见于Excel和类似的表格应用中。数据透视表将每一列的数据作为输入，输出将数据不断细分成多个维度累计信息的二维数据表。人们有时容易弄混数据透视表和GroupBy，但我觉得数据透视表更像是一种 **多维** 的GroupBy累计操作。也就是说，虽然你也可以分割-应用-组合，但是分割与组合不是发生在一维索引上，而是二维网格上（行列同时分组）。

### 演示数据透视表

这一节的示例用泰坦尼克号的乘客信息数据库来演示，可以在Seaborn程序中获取：

````python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
````

````python
titanic.head()
````

### 手工制作数据透视表

我们想将它们按照性别、最终生还状态或其他组合属性进行分组。你可能会用GroupBy来实现，例如这样统计不同性别乘客的生还率。

````python
titanic.groupby('sex')['survived'].mean()
````

````python
titanic.groupby('sex')[['survived']].mean()
````

上面有些不同，为什么？

总体来说，有四分之三的女生被拯救，但只有五分之一的男性被救。

我们还想进一步探索，同时观察不同性别与船舱等级的生还情况。根据GroupBy的操作流程，我们也许能实现想要的结果：将class和sex分组，然后选择生还列，应用均值累计函数，再将各组结果组合。最后通过行索引转列索引操作，将最里层的行索引转换成列索引，形成二维数组。代码如下：

````python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
````

上面那段代码不太方便，而且GroupBy的应用场景非常普遍，所以Pandas提供了一个快捷方式pivot_table来快速解决多维的累计分析任务。

### 数据透视表语法

用DataFrame的pivot_table实现的效果等同于上一节的管道命令的代码：

````python
titanic.pivot_table('survived', index='sex', columns='class')
````

````python
titanic.pivot_table?
````

#### 多级数据透视表

与GroupBy类似，数据透视表中的分组也可以通过各种参数指定多个等级。例如，我们想把年龄也加进去作为第三个维度，这就可以通过pd.cut函数将年龄进行分段：

````python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
````

对某一列也可以使用同样的策略——让我们用pd.qcut将船票价格按照计数项等分为2份，加入数据透视表看看：

````python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
````

结果是一个带层级索引的四维累计数据表，通过网格显示不同数值之间的相关性。

#### 其他数据透视表选项

DataFrame.pivot_table的完整签名如下：

````python
titanic.pivot_table?
````

前三个参数已经介绍过了，现在看看其他参数。fill_value和dropna这两个参数用于处理缺失值。

aggfunc参数用于设置累计函数类型，默认值是均值(mean)。与GroupBy的用法一样，累计函数可以用一些常见的字符串（'sum', 'mean', 'count', 'min', 'max'等）表示，也可以用标准的累计函数（np.sum(), min(), sum()等）表示。另外，还可以通过字典为不同的列指定不同的累计函数：

````python
titanic.pivot_table(index='sex', columns='class',
                    aggfunc={'survived':sum, 'fare':'mean'})
````

需要注意的是，这里忽略了一个参数values。这是因为当我们为aggfunc指定映射关系的时候，待透视的数值就已经确定了。

当需要计算每一组的总数时，可以通过margins参数来设置：

````python
titanic.pivot_table('survived', index='sex', columns='class', margins=True)
````

### 案例：美国人的生日

你可以下载公开的生日数据。并导入：

````python
import pandas as pd
import numpy as np
births = pd.read_csv('data//births.csv')
births.head()
````

可以用一个数据透视表来探索这份数据，先增加一列表示不同年代，看看各年代的男女出生比例：

````python
births['decade'] = 10 * (births['year'] // 10)
# births['decade'] = births['year'] # 上面那个写法是表示年代的意思
births.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
````

我们可以马上发现，每个年代的男性出生率都比女性出生率高。如果希望更直观地体现这种趋势，可以用Pandas内置的画图功能将每一年的出生人数画出来。

````python
%matplotlib inline
import matplotlib.pyplot as plt
sns.set()   # 使用Seaborn风格
births.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('total births per year');
````

使用Pandas工具会展示一些有趣的特征。首先我们要做一点儿清理工作，消除由于输错日期而造成的异常点（如6月31日）或者是缺失值（如1999年6月）。消除这些异常的简便方法就是直接删除异常点，可以通过稳健的sigma消除法（sigma-clipping，按照正态分布标准划定范围，SciPy中默认是四个标准差）操作来实现：

````python
quartiles = np.percentile(births['births'], [25, 50, 75])
mu = quartiles[1]
sig = 0.74 * (quartiles[2] - quartiles[0])
````

最后一行是样本均值的稳定性估计，其中0.74是指标准正态分布的分位数间距。在query方法中用这个范围就可以将有效的生日数据筛选出来：

````python
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
````

然后，将day列设置为整数。这列数据在筛选之前是字符串，因为数据集中有的列含有缺失值null：

````python
# 将day列设置成整数。
births['day'] = births['day'].astype(int)
````

````python
births['day'].head()
````

现在可以将年月日组合起来创建一个日期索引了，这样可以快速计算每一行是星期几：

````python
# 从年月日创建一个日期索引
births.index = pd.to_datetime(10000 * births.year +
                              100 * births.month +
                              births.day, format='%Y%m%d')
births['dayofweek'] = births.index.dayofweek
````

用这个索引可以画出不同年代不同星期的日均出生数据：

````python
import matplotlib.pyplot as plt
import matplotlib as mpl

births.pivot_table('births', index='dayofweek',
                   columns='decade', aggfunc='mean').plot()
plt.gca().set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('mean births by day');
````

由图可知，周末的出生人数比工作日的要低很多。另外，因为CDC只提供了1989年前的数据，所以没有20世纪90年代和21世纪的数据。
